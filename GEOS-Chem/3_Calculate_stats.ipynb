```python
from scipy.stats import pearsonr
def calR(mo):
    corrData = pd.DataFrame(columns=['lat','lon','r_AK','r_without','season','count','slope','p','R2','RMSE','p_r','p_r_without',
                                     'slope_without','p_without','R2_without','RMSE2_without'])
# for mo in list(monthlyData['Season'].unique()):
# for mo in [1]:

#     print(mo)
    temp = monthlyData[monthlyData['Season']==mo]
#     print(str(len(temp['lat']))+"!!!")
    
#     for hr in list(temp['hour'].unique()):
#         temp2 = temp[temp['hour']==hr]
        
#         for lat in list(temp2['latX'].unique()):
    for lat in Vlist:
        for lon in Hlist:

            temp2 = temp[(temp['lat']==lat) & (temp['lon']==lon)]
            temp2 = temp2.dropna()
            if(len(temp2['lat'])>1):
                rvalue,p1 = pearsonr(temp2['VCD_AK'],temp2['HCHO_column'])
                rvalue2,p2 = pearsonr(temp2['VCD'],temp2['HCHO_column'])
#                 rvalue = temp2['VCD_AK'].corr(temp2['HCHO_column'])
#                 rvalue2 = temp2['VCD'].corr(temp2['HCHO_column'])
#                 print(str(lat)+' '+str(lon)+ ' '+str(hr)+' '+str(rvalue))
                
#                 reg = LinearRegression()
#                 reg.fit(temp2['VCD_AK'], temp2['HCHO_column'])
#                 slope = reg.coef_[0]
#                 r2 = reg.score(temp2['VCD_AK'], temp2['HCHO_column'])
                result = smf.ols(formula='VCD_AK ~ HCHO_column - 1', data = temp2).fit()
                slope , p = result.params['HCHO_column'], result.pvalues['HCHO_column']
                y_pred = result.predict(temp2['HCHO_column'])
                rmse2 = rmse(temp2['VCD_AK'],y_pred)
                r2 = result.rsquared
                
                result2 = smf.ols(formula='VCD ~ HCHO_column - 1', data = temp2).fit()
                slope2 , pp = result2.params['HCHO_column'], result2.pvalues['HCHO_column']
                y_pred2 = result2.predict(temp2['HCHO_column'])
                rmse22 = rmse(temp2['VCD'],y_pred2)
                r22 = result2.rsquared
#                 print(str(slope)+' '+str(p)+ ' '+str(r2)+ ' '+str(rmse2))
                corrData.loc[len(corrData.index)] = [lat, lon, rvalue, rvalue2, mo, len(temp2['lat'])
                                                     , slope, p, r2,rmse2,p1,p2, slope2, pp, r22,rmse22]
    corrData.to_pickle('/media/dell/Elements/HCHO_GEMS_V2/CorrelationGC_GEMS/SeasonalGEMS_GC_R3_'+mo)
                
                
                
#                 print(str(len(temp3['latX']))+"!!")
                    
#                 print(str(len(temp3['latX']))+"!")
            
                
#                 print(str(len(temp3['latX']))+"?")
#             if(len(temp['lat'])>4):
                    
    
#             else:
#                 corrData.loc[len(corrData.index)] = [lat, lon, 0, monID]
# corrData
```
